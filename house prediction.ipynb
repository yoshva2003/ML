```python
x=hd[['bedrooms', 'bathrooms', 'sqft_living']]
y=hd['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.45,random_state=110)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
coeff_hd=pd.DataFrame(lr.coef_,x.columns,columns=['coefficient'])
```

```python
coeff_hd
```

```python
predict=lr.predict(x_test)
```

```python
plt.scatter(y_test,predict)
```

```python
sns.distplot((y_test-predict),bins=50)
```
