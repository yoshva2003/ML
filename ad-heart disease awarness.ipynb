```python
import pandas as pd
```

```python
data=pd.read_csv('heart-disease.csv')
```

```python
data.head()
```

```python

import matplotlib.pyplot as plt

# Line Chart
plt.figure(figsize=(10, 6))
plt.plot(data["Year"], data["India_Deaths_Millions"], label="India", marker='o')
plt.plot(data["Year"], data["World_Deaths_Millions"], label="World", marker='o')
plt.title("Heart Disease Deaths Over Time", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Deaths (in Millions)", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data["Year"], data["India_Contribution_Percentage"], color='teal')
plt.title("India's Contribution to Global Heart Disease Deaths (%)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Percentage", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data["India_Population_Millions"], data["India_Deaths_Millions"], color='crimson', alpha=0.7)
plt.title("India: Population vs. Heart Disease Deaths", fontsize=16)
plt.xlabel("Population (in Millions)", fontsize=12)
plt.ylabel("Deaths (in Millions)", fontsize=12)
plt.grid(True)
plt.show()
```

```python
year_filter = data[data["Year"] == 2020]
india_deaths = year_filter["India_Deaths_Millions"].values[0]
world_deaths = year_filter["World_Deaths_Millions"].values[0] - india_deaths

# Pie Chart
labels = ['India', 'Rest of the World']
sizes = [india_deaths, world_deaths]
colors = ['red', 'gray']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Heart Disease Deaths: India vs. Rest of the World (2020)")
plt.show()
```

```python
import seaborn as sns

# Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Heart Disease Data", fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(12, 6))
plt.plot(data["Year"], data["India_Deaths_Millions"], label="India", color="red", linewidth=3, marker='o')
plt.plot(data["Year"], data["World_Deaths_Millions"], label="World", color="blue", linewidth=3, linestyle="--", marker='s')
plt.title("Rising Heart Disease Deaths Over Time", fontsize=20, fontweight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Deaths (in Millions)", fontsize=14)
plt.legend(fontsize=12, loc="upper left")
plt.grid(visible=True, which='major', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("line_chart_ad.png", dpi=300)  # Save the chart for ad use
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(data["Year"], data["India_Contribution_Percentage"], color='orange', edgecolor='black')
plt.title("India's Growing Contribution to Global Heart Disease Deaths", fontsize=20, fontweight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Percentage (%)", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("bar_chart_ad.png", dpi=300)  # Save the chart for ad use
plt.show()
```

```python

# Plot the trend chart
plt.figure(figsize=(12, 6))
plt.plot(data["Year"], data["India_Deaths_Millions"], label="India", color="red", linewidth=3, marker='o')
plt.plot(data["Year"], data["World_Deaths_Millions"], label="World", color="blue", linewidth=3, linestyle="--", marker='s')

# Title and labels
plt.title("Rising Heart Disease Deaths Over Time", fontsize=20, fontweight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Deaths (in Millions)", fontsize=14)

# Legend and grid
plt.legend(fontsize=12, loc="upper left")
plt.grid(visible=True, which='major', linestyle='--', alpha=0.6)

# Save and show the chart
plt.tight_layout()
plt.savefig("line_chart_ad.png", dpi=300)  # Save the chart as a PNG file
plt.show()
```

```python

# Prepare data for the stacked area chart
years = data["Year"]
india_deaths = data["India_Deaths_Millions"]
world_minus_india_deaths = data["World_Deaths_Millions"] - india_deaths

# Plot the stacked area chart
plt.figure(figsize=(12, 6))
plt.fill_between(years, 0, world_minus_india_deaths, label="Rest of the World", color="lightblue", alpha=0.8)
plt.fill_between(years, world_minus_india_deaths, world_minus_india_deaths + india_deaths, 
                 label="India", color="orange", alpha=0.8)

# Add title and labels
plt.title("Heart Disease Death Trends: India vs. Rest of the World", fontsize=20, fontweight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Deaths (in Millions)", fontsize=14)

# Add legend and grid
plt.legend(fontsize=12, loc="upper left")
plt.grid(visible=True, which='major', linestyle='--', alpha=0.6)

# Save and show the chart
plt.tight_layout()
plt.savefig("stacked_area_chart_ad.png", dpi=300)  # Save the chart for ad use
plt.show()
```

```python

# Prepare data for the bubble chart
years = data["Year"]
population = data["India_Population_Millions"]
deaths = data["India_Deaths_Millions"]
contribution_percentage = data["India_Contribution_Percentage"]

# Plot the bubble chart
plt.figure(figsize=(12, 6))
bubble_sizes = contribution_percentage * 10  # Scale bubble sizes for better visualization
scatter = plt.scatter(population, deaths, s=bubble_sizes, c=years, cmap="viridis", alpha=0.7, edgecolors="black")

# Add color bar to represent years
cbar = plt.colorbar(scatter)
cbar.set_label("Year", fontsize=12)

# Add title and labels
plt.title("Population vs. Heart Disease Deaths in India", fontsize=20, fontweight='bold')
plt.xlabel("Population (in Millions)", fontsize=14)
plt.ylabel("Deaths (in Millions)", fontsize=14)

# Add grid for better readability
plt.grid(True, alpha=0.5, linestyle='--')

# Save and show the chart
plt.tight_layout()
plt.savefig("bubble_chart_ad.png", dpi=300)  # Save the chart for ad use
plt.show()
```
